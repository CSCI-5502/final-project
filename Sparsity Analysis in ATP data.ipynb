# Other Statistics from 2001 to 2016

```python
series_summary = df['Series'].value_counts()
print(series_summary)
```

```python
tourney_summary = df['Tournament'].value_counts()
print(tourney_summary)
```

```python
winner_summary = df['Winner'].value_counts()
print(winner_summary)
```

```python
loser_summary = df['Loser'].value_counts()
print(loser_summary)
```

```python
all_match_players = pd.concat([df['Winner'], df['Loser']])
print(all_match_players.shape)  
print(df.shape[0] * 2)
```

```python
#list of players
players = []
players = all_match_players.unique()
print(len(players))
```

```python
print("Player \t Number of matches from 2001 to 2016")
player_match_count = all_match_players.value_counts()
print(player_match_count)
```

```python
print("NUMBER OF PLAYERS WITH MORE THAN ONE MATCHES \n \n")
for i in range(1260):
    if player_match_count[players[i]] > 1:
        print('{:<30s} {:<10d}'.format(players[i], player_match_count[players[i]]))
```
