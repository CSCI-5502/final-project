# Sparsity statistics from 2003-2016

Since the number of attrtibutes have increased from 30 to 49 from the year 2003, I am concatinating the data for the years 2003-2016 for further analysis.

```python
df2003 = pd.read_csv('wta_matches_2003.csv')
df2004 = pd.read_csv('wta_matches_2004.csv')
df2005 = pd.read_csv('wta_matches_2005.csv')
df2006 = pd.read_csv('wta_matches_2006.csv')
df2007 = pd.read_csv('wta_matches_2007.csv')
df2008 = pd.read_csv('wta_matches_2008.csv')
df2009 = pd.read_csv('wta_matches_2009.csv')
df2010 = pd.read_csv('wta_matches_2010.csv')
df2011 = pd.read_csv('wta_matches_2011.csv')
df2012 = pd.read_csv('wta_matches_2012.csv', encoding = "ISO-8859-1")
df2013 = pd.read_csv('wta_matches_2013.csv')
df2014 = pd.read_csv('wta_matches_2014.csv')
df2015 = pd.read_csv('wta_matches_2015.csv')
df2016 = pd.read_csv('wta_matches_2016.csv')

frames = [df2003, df2004, df2005, df2006, df2007, df2008, df2009, df2010, df2011, df2012, df2013, df2014, df2015, df2016] 
df = pd.concat(frames, sort = 'False')
display(df)
```

```python
#list of players
players = []
players = all_match_players.unique()
print(len(players))
```

```python
print("Player \t Number of matches in 2003-2014;2016")
player_match_count = all_match_players.value_counts()
print(player_match_count)
```

```python
print("NUMBER OF PLAYERS WITH MORE THAN ONE MATCHES \n \n")
for i in range(1591):
    if player_match_count[players[i]] > 1:
        print('{:<30s} {:<10d}'.format(players[i], player_match_count[players[i]])) 
        
```
